```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
import astropy.constants as c

def BB(nu,temp):
    intensity = 2*c.h*nu**3 / c.c**2 / (np.exp(c.h*nu/c.k_B/temp)-1)
    return intensity.to(u.erg/u.s/u.Hz/u.cm**2)
freqb3 = 92982346121.91989*u.Hz
freqb6 = 226382496476.07114*u.Hz
temparr = np.linspace(5,200,100)*u.K
alphaarr = []
for temp in temparr:
    B3flux = BB(freqb3,temp) 
    B6flux = BB(freqb6,temp)
    alpha = np.log10(B3flux/B6flux)/np.log10(freqb3/freqb6)
    alphaarr.append(alpha)
alphaarr = np.array(alphaarr)
fig = plt.figure(figsize=(8,8))
plt.plot(temparr, alphaarr)
plt.ylabel('spectral index')
plt.xlabel('temperature (K)')
print(np.argmin(np.abs(alphaarr-1.7)))
print(temparr[np.argmin(np.abs(alphaarr-1.7))])
```
