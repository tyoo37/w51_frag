```python
import csv
import numpy as np
import matplotlib.pyplot as plt
dat = np.loadtxt(open("fragmentation_table.csv", "rb"), delimiter=",", skiprows=1)
scale_tab = dat[:,0]
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
                 
for i in range(9):
    ax1.plot(scale_tab, dat[:,i+1])
print(dat.shape)
ax1.set_xscale('log')
ax1.set_xlim(1e-5,1e-1)

ax1.set_ylabel('fragmentation rate')
ax1.set_xlabel('R (pc)')

```
