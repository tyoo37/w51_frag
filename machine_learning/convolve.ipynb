```python
import numpy as np
from sedfitter.filter import Filter
from sedfitter.convolve import convolve_model_dir
from astropy import units as u
from astroquery.svo_fps import SvoFps
import glob
import os
seddir = '/orange/adamginsburg/w51/TaehwaYoo/SEDmodel_new/'
pathlist = glob.glob('/blue/adamginsburg/richardson.t/research/flux/r+24_models-1.2/s*')
geometry = [os.path.basename(path).split('/')[-1][:7] for path in pathlist]

almaimf_bandends_1mm = [[216.10085679, 216.36181569],
                        [217.05104378, 217.31175857],
                        [219.90488464, 220.04866835],
                        [218.13102322, 218.39222624],
                        [219.51976276, 219.66379059],
                        [230.31532951, 230.81137113],
                        [231.06503709, 231.56181105],
                        [231.52507012, 233.42623749]]*u.GHz

nu_1mm = np.linspace(almaimf_bandends_1mm.min(), almaimf_bandends_1mm.max(), 5000)
response_1mm = np.zeros(nu_1mm.size, dtype='bool')
for start, stop in almaimf_bandends_1mm:
    response_1mm |= (nu_1mm > start) & (nu_1mm < stop)
    
almaimf_bandends_3mm = [[ 93.13410936,  93.25141259],
                        [ 91.75059068,  92.68755174],
                        [102.15273354, 103.0896946 ],
                        [104.55323851, 105.49019957]]*u.GHz

nu_3mm = np.linspace(almaimf_bandends_3mm.min(), almaimf_bandends_3mm.max(), 5000)
response_3mm = np.zeros(nu_3mm.size, dtype='bool')

for start, stop in almaimf_bandends_3mm:
    response_3mm |= (nu_3mm > start) & (nu_3mm < stop)
"""
band7_band = [[336.5-1.875/2, 336.5+1.875/2],
              [338.5-1.875/2, 338.5+1.875/2],
              [348.5-1.875/2, 348.5+1.875/2],
              [350.5-1.875/2, 350.5+1.875/2]]*u.GHz

nu_band7 = np.linspace(band7_band.min(), band7_band.max(),5000)
response_band7 = np.zeros(nu_band7.size, dtype='bool')
for start, stop in band7_band:
    response_band7 |= (nu_band7>start) & (nu_band7<stop)
"""
filternames = ['user_filters/ALMA-IMF_band3', 'user_filters/ALMA-IMF_band6']
nus = [nu_3mm, nu_1mm]
responses = [response_3mm.astype('float'), response_1mm.astype('float')]
wavelengths = [(99.68314596*u.GHz).to(u.mm, u.spectral()),(228.15802*u.GHz).to(u.mm, u.spectral())]
sed_filters = [Filter(name=filtername,central_wavelength=wl,
                      nu=nu,response=response)
                   for filtername, wl, nu,response in zip(filternames, wavelengths, nus, responses)]

#sed_filters = [Filter(name='user_filters/ALMA-IMF_band7', central_wavelength=(343.5*u.GHz).to(u.mm, u.spectral()), 
#                    nu=nu_band7, response=response_band7.astype(float))]
for sed_filter in sed_filters:
    sed_filter.normalize()
for model_dir in geometry:
    try:
        os.system('mkdir -p '+seddir+model_dir+'/convolved/user_filters')
        print('mkdir -p '+seddir+model_dir+'/convolved/user_filters')
    except:
        raise ValueError
    #os.system('mkdir '+seddir+model_dir+'convolved')
    #os.system('mkdir '+seddir+model_dir+'convolved/user_filters')

    if not os.path.isfile(seddir+model_dir+'flux.fits'):
        os.system('cp /blue/adamginsburg/richardson.t/research/flux/r+24_models-1.2/%s/*.fits %s'%(model_dir,seddir+model_dir)) 
    convolve_model_dir(seddir+model_dir,sed_filters,overwrite=True)
    print(seddir+model_dir)





```
