```python
from astropy.table import Table

w51e_dendro_matched_catalog = '/home/t.yoo/w51/catalogue/dendrogram/dendro_w51e_matched.fits'
w51n_dendro_matched_catalog = '/home/t.yoo/w51/catalogue/dendrogram/dendro_w51n_matched.fits'

photometrydir = '/home/t.yoo/w51/catalogue/photometry/flux_new/'
w51e_b3_flux_file = photometrydir+'w51e_b3_test.fits'
w51n_b3_flux_file = photometrydir+'w51n_b3_test.fits'
w51e_b6_flux_file = photometrydir+'w51e_b6_test.fits'
w51n_b6_flux_file = photometrydir+'w51n_b6_test.fits'
w51e_b6_conv_flux_file = photometrydir+'w51e_b6_conv_test.fits'
w51n_b6_conv_flux_file = photometrydir+'w51n_b6_conv_test.fits'



```

```python
w51e_dendro_matched= Table.read(w51e_dendro_matched_catalog, format='fits')
w51e_b3_flux = Table.read(w51e_b3_flux_file, format='fits')
print(w51e_b3_flux)
for column in w51e_b3_flux.colnames:
    print(column)
    w51e_dendro_matched[column+'_b3'] = w51e_b3_flux[column]

w51e_b6_flux = Table.read(w51e_b6_flux_file, format='fits')
for column in w51e_b6_flux.colnames:
    w51e_dendro_matched[column+'_b6'] = w51e_b6_flux[column]
    
w51e_b6_conv_flux = Table.read(w51e_b6_conv_flux_file, format='fits')
for column in w51e_b6_conv_flux.colnames:
    w51e_dendro_matched[column+'_b6_conv'] = w51e_b6_conv_flux[column]    
w51e_dendro_matched.write(w51e_dendro_matched_catalog, format='fits', overwrite=True)
```

```python

w51n_dendro_matched= Table.read(w51n_dendro_matched_catalog, format='fits')
w51n_b3_flux = Table.read(w51n_b3_flux_file, format='fits')
for column in w51n_b3_flux.colnames:
    w51n_dendro_matched[column+'_b3'] = w51n_b3_flux[column]

w51n_b6_flux = Table.read(w51n_b6_flux_file, format='fits')
for column in w51n_b6_flux.colnames:
    w51n_dendro_matched[column+'_b6'] = w51n_b6_flux[column]

w51n_b6_conv_flux = Table.read(w51n_b6_conv_flux_file, format='fits')
for column in w51n_b6_conv_flux.colnames:
    w51n_dendro_matched[column+'_b6_conv'] = w51n_b6_conv_flux[column]   
    
w51n_dendro_matched.write(w51n_dendro_matched_catalog, format='fits', overwrite=True)
```
